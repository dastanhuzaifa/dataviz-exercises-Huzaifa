## Task 1 — Histogram: price by room type (overlapping distributions)

**What to build:** A histogram showing price distributions for **Entire home/apt vs Private room** (exclude Shared room — too few observations) overlaid on the same chart.

**Requirements:**
- Both room types on the same chart (use `color='room_type'`)
- `barmode='overlay'` with `opacity=0.6` so both distributions are visible
- A vertical line for the median of EACH room type, differently coloured
- Insight title comparing the two distributions

> 💡 `df_cap[df_cap['room_type'].isin(['Entire home/apt','Private room'])]`


```python
import pandas as pd
import plotly.express as px
import numpy as np

# Load dataset
df = pd.read_csv('../data/airbnb_london.csv')

# Cap price outliers at 95th percentile
p95 = df['price'].quantile(0.95)
df_cap = df[df['price'] <= p95].copy()

print(f"95th percentile price cap: £{p95:.0f}")

# Filter to the two main room types
room_df = df_cap[
    df_cap['room_type'].isin(['Entire home/apt', 'Private room'])
]

# Calculate medians
median_entire = room_df.loc[
    room_df['room_type'] == 'Entire home/apt',
    'price'
].median()

median_private = room_df.loc[
    room_df['room_type'] == 'Private room',
    'price'
].median()

# Build histogram
fig = px.histogram(
    room_df,
    x='price',
    color='room_type',
    barmode='overlay',
    opacity=0.6,
    nbins=40,
    color_discrete_map={
        'Entire home/apt': '#0072B2',
        'Private room': '#E69F00'
    },
    title='Entire Homes Have a Wider and Higher-Priced Distribution Than Private Rooms'
)

# Add median lines
fig.add_vline(
    x=median_entire,
    line_dash='dash',
    line_color='#0072B2',
    annotation_text=f'Entire home median: £{median_entire:.0f}',
    annotation_position='top'
)

fig.add_vline(
    x=median_private,
    line_dash='dash',
    line_color='#E69F00',
    annotation_text=f'Private room median: £{median_private:.0f}',
    annotation_position='top'
)

# Layout updates
fig.update_layout(
    xaxis_title='Nightly Price (£)',
    yaxis_title='Number of Listings',
    bargap=0.05,
    margin=dict(t=70, l=10, r=10, b=10)
)

fig.show()
```

## Task 2 — Box plot: listing activity by borough

**What to build:** A **horizontal box plot** comparing listing activity (reviews per month) across London boroughs — reviews per month is a proxy for how frequently a listing is booked.

**Requirements:**
- Horizontal orientation (borough names are long)
- Sorted by median reviews per month (most active at top)
- Highlight the **two most active** boroughs in a different colour
- Outliers shown as individual points
- Insight title naming the two busiest boroughs

> 💡 Some listings have zero reviews — these are new or inactive listings. Filter them out with before plotting

```python
import pandas as pd
import plotly.express as px
import numpy as np

# Load dataset
df = pd.read_csv('../data/airbnb_london.csv')

# Check columns if needed
print(df.columns)

# Filter out inactive listings
active_df = df[df['reviews_per_month'] > 0].copy()

# Median reviews per month by borough/neighbourhood
borough_medians = (
    active_df.groupby('neighbourhood')['reviews_per_month']
    .median()
    .sort_values(ascending=False)
)

# Order boroughs by median activity
borough_order = borough_medians.index.tolist()

# Identify top 2 busiest boroughs
top_two = borough_order[:2]

# Create highlight category
active_df['highlight'] = np.where(
    active_df['neighbourhood'].isin(top_two),
    'Top boroughs',
    'Other boroughs'
)

# Build horizontal box plot
fig = px.box(
    active_df,
    x='reviews_per_month',
    y='neighbourhood',
    orientation='h',
    color='highlight',
    points='outliers',
    category_orders={
        'neighbourhood': borough_order[::-1]
    },
    color_discrete_map={
        'Top boroughs': '#D55E00',   # orange-red
        'Other boroughs': '#999999'  # grey
    },
    title=f'{top_two[0]} and {top_two[1]} Show the Highest Airbnb Listing Activity'
)

# Add overall median reference line
overall_median = active_df['reviews_per_month'].median()

fig.add_vline(
    x=overall_median,
    line_dash='dash',
    line_color='black',
    annotation_text=f'Overall median: {overall_median:.1f}',
    annotation_position='top'
)

# Layout updates
fig.update_layout(
    xaxis_title='Reviews per Month',
    yaxis_title='London Borough',
    showlegend=False,
    margin=dict(t=70, l=10, r=10, b=10)
)

fig.show()
```
